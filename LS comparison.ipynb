```python
import numpy as np
import scipy
import gatspy
import matplotlib.pyplot as plot
from gatspy.periodic import LombScargleFast
import astropy
from astropy.timeseries import LombScargle
```

```python
#Comparing LombScargle and LombScargle fast
def test_lc(t):
    return np.sin((2*np.pi*t)/10)

t = np.linspace(0,50,1000)
y = test_lc(t)

#LSFast
model_f = LombScargleFast(silence_warnings=True).fit(t,test_lc(t))
model_f.optimizer.period_range = (1, 20)
periods, power = model_f.periodogram_auto(nyquist_factor=2000)
print("LombScargleFast:",model_f.best_period)

#LS 
frequency, power_astropy = LombScargle(t, y).autopower(
    minimum_frequency=1/20,
    maximum_frequency=1/1,
    samples_per_peak=20)

periods_astropy = 1 / frequency
best_period_astropy = periods_astropy[np.argmax(power_astropy)]
print("LombScargle:", best_period_astropy)
print("True Period: 10")
```
